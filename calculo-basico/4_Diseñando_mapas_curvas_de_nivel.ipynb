## Dependences

```python
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt
```

## Grafica en 3 dimensiones

```python
def f(x, y):
    return np.sin(x) + 2 * np.cos(y)

res = 100
x = np.linspace(-4.0, 4.0, num=res)
y = np.linspace(-4.0, 4.0, num=res)

x, y = np.meshgrid(x, y)

z = f(x, y)

fig, ax = plt.subplots(subplot_kw={'projection': '3d'})

surf = ax.plot_surface(x, y, z, cmap=cm.cool)
fig.colorbar(surf)
```

## Curvas de nivel

```python
fig2, ax2 = plt.subplots()
level_map = np.linspace(np.min(z), np.max(z), num=res//2)

cp = ax2.contour(x, y, z, levels=level_map, cmap=cm.cool)
fig2.colorbar(cp)
```
